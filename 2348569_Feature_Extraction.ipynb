```python
point_mask = np.array([[1, 1, 1],
                       [1, -8, 1],
                       [1, 1, 1]])
img_point1 = cv2.filter2D(img1, -1, point_mask)
img_point2 = cv2.filter2D(img2, -1, point_mask)
img_point3 = cv2.filter2D(img3, -1, point_mask)
img_point4 = cv2.filter2D(img4, -1, point_mask)
```

# **Line Extraction**

```python
vertical_mask = np.array([[-1, 2, -1],
                          [-1, 2, -1],
                          [-1, 2, -1]])
horizontal_mask = np.array([[-1, -1, -1],
                            [2, 2, 2],
                            [-1, -1, -1]])

plus_45_mask = np.array([[-1, -1, 2],
                         [-1, 2, -1],
                         [2, -1, -1]])

minus_45_mask = np.array([[2, -1, -1],
                          [-1, 2, -1],
                          [-1, -1, 2]])
```

```python
#image 3
vertical_lines3 = cv2.filter2D(img3, -1, vertical_mask)
horizontal_lines3 = cv2.filter2D(img3, -1, horizontal_mask)
plus_45_lines3 = cv2.filter2D(img3, -1, plus_45_mask)
minus_45_lines3 = cv2.filter2D(img3, -1, minus_45_mask)
```

```python
#image 4
vertical_lines4 = cv2.filter2D(img4, -1, vertical_mask)
horizontal_lines4 = cv2.filter2D(img4, -1, horizontal_mask)
plus_45_lines4 = cv2.filter2D(img4, -1, plus_45_mask)
minus_45_lines4 = cv2.filter2D(img4, -1, minus_45_mask)
```

# **Edge Extraction**

```python
def detect_edges(image):
    edges = cv2.Canny(image, 100, 200)
    return edges
```

```python
edges_image1 = detect_edges(img1)
edges_image2 = detect_edges(img2)
edges_image3 = detect_edges(img3)
edges_image4 = detect_edges(img4)
```

```python
plt.figure(figsize=(14,12))

plt.subplot(1,4,1)
plt.title("Edges in Image 1")
plt.imshow(edges_image1,cmap='gray')
plt.axis("off")

plt.subplot(1,4,2)
plt.title("Edges in Image 2")
plt.imshow(edges_image2,cmap='gray')
plt.axis("off")

plt.subplot(1,4,3)
plt.title("Edges in Image 3")
plt.imshow(edges_image3,cmap='gray')
plt.axis("off")

plt.subplot(1,4,4)
plt.title("Edges in Image 4")
plt.imshow(edges_image4,cmap='gray')
plt.axis("off")

plt.show()
```
