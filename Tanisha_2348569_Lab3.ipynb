```python
import cv2
from google.colab.patches import cv2_imshow

# Read the image
image = cv2.imread('/content/drive/MyDrive/IMAGES (CV)/angel.jpg')
print('Image:')
cv2_imshow(image)
```

# **Image Negative**

The image negative is a simple image enhancement technique that involves reversing the intensity values of pixels in an image. For grayscale images, the negative of each pixel intensity is computed as 255 - intensity_value. For color images, the negative of each color channel is computed independently.
This technique is often used to enhance the contrast of an image by making dark areas lighter and vice versa.


```python
# Compute the negative of the image
negative_image = 255 - image

# Display or save the negative image
print('Negative Image')
cv2_imshow(negative_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# **Gray Level Slicing**

Gray level slicing is a technique used to enhance specific ranges of pixel intensities in an image. It involves setting the intensity values of pixels within a certain range to a specific value (e.g., white) while leaving other pixels unchanged or setting them to another value (e.g., black). This technique is useful for highlighting or enhancing specific features or regions of interest in an image by increasing their visibility against the background.

```python
import numpy as np

def gray_level_slicing(image, lower_thresh, upper_thresh):
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply gray level slicing
    output_image = np.copy(image)
    output_image[(gray_image >= lower_thresh) & (gray_image <= upper_thresh)] = [255, 255, 255]  # Set to white
    output_image[(gray_image < lower_thresh) | (gray_image > upper_thresh)] = [0, 0, 0]           # Set to black

    return output_image

# Apply gray level slicing
sliced_image = gray_level_slicing(image, 100, 200)

# Display or save the sliced image
print('Sliced Image')
cv2_imshow(sliced_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# **Histogram Equalization**

Histogram equalization is a method used to improve the contrast of an image by redistributing the intensity values of pixels. It works by transforming the intensity histogram of the image such that the cumulative distribution function (CDF) of pixel intensities becomes approximately uniform. This technique effectively spreads out the intensity values across the entire dynamic range, making both dark and bright regions more distinguishable.
Histogram equalization is particularly useful for enhancing images with low contrast or uneven lighting conditions.

```python
img = cv2.imread('/content/drive/MyDrive/IMAGES (CV)/angel.jpg', cv2.IMREAD_GRAYSCALE)

# Apply histogram equalization
equalized_image = cv2.equalizeHist(img)

# Display or save the equalized image
print('Equalized Image')
cv2_imshow(equalized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
if image is None:
    print("Error: Unable to read the image.")
else:
    # Split the image into its color channels
    b, g, r = cv2.split(image)

    # Apply histogram equalization to each color channel
    b_equalized = cv2.equalizeHist(b)
    g_equalized = cv2.equalizeHist(g)
    r_equalized = cv2.equalizeHist(r)

    # Merge the equalized color channels back into an image
    equalized_image = cv2.merge([b_equalized, g_equalized, r_equalized])

    # Display or save the equalized image
    print('Equalized color Image: ')
    cv2_imshow(equalized_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
