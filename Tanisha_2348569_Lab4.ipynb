# **Linear Spatial Filtering using Cv2 (Built-in Functions)**

```python
# Define the kernel for filtering
kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])
```

```python
# Apply the filter using OpenCV
filtered_image = cv2.filter2D(image, -1, kernel)
```

```python
# Display the original and filtered images
print('Filtered Image:')
cv2_imshow(filtered_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# **Linear Spatial Filtering using user-defined functions**

```python
def linear_filter(image, kernel):
    # Apply the filter using user-defined function
    rows, cols, channels = image.shape
    kernel_size = len(kernel)
    pad_size = kernel_size // 2
    filtered_image = np.zeros_like(image)

    for i in range(pad_size, rows - pad_size):
        for j in range(pad_size, cols - pad_size):
            roi = image[i-pad_size:i+pad_size+1, j-pad_size:j+pad_size+1]
            for c in range(channels):
                filtered_image[i, j, c] = np.sum(roi[:,:,c] * kernel)

    return filtered_image
```
